```python
new_df = df.pivot(index="date", columns="country", values="people_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=new_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```

```python
new_df = df.pivot(index="date", columns="country", values="people_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
top10_df = new_df[new_df[df['country'].drop_duplicates()].max().sort_values(ascending=False)[:10].index]

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=top10_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```

```python
new_df = df.pivot(index="date", columns="country", values="people_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
new_df.loc[:,new_df[df['country'].drop_duplicates()].max() > 0].max().sort_values()[:10].index

least10_df = new_df[new_df.loc[:,new_df[df['country'].drop_duplicates()].max() > 0].max().sort_values()[:10].index]

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=least10_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```
