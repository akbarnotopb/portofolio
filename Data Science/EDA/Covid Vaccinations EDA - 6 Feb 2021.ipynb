```python
new_df = df.pivot(index="date", columns="country", values="people_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=new_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```

```python
new_df = df.pivot(index="date", columns="country", values="people_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
top10_df = new_df[new_df[df['country'].drop_duplicates()].max().sort_values(ascending=False)[:10].index]

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=top10_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```

```python
new_df = df.pivot(index="date", columns="country", values="people_fully_vaccinated")
new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
new_df.loc[:,new_df[df['country'].drop_duplicates()].max() > 0].max().sort_values()[:10].index

least10_df = new_df[new_df.loc[:,new_df[df['country'].drop_duplicates()].max() > 0].max().sort_values()[:10].index]

plt.figure(figsize=(15, 8))
ax = sns.lineplot(data=least10_df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)
ax.xaxis.set_major_locator(locator)
ax.set_ylabel("people vaccinated")
```

# Daily Vaccination

```python
from sklearn import preprocessing

type_accines = df.pivot(index="date", columns="country", values="vaccines")
# new_df = new_df.interpolate(method="time")
type_accines.fillna("Unknown", inplace = True)
type_accines = type_accines[["United States", "China", "Luxembourg", "Iceland"]]
observed = ["United States", "China", "Luxembourg", "Iceland"]
for  ob in observed:
    type_accines[ob] = type_accines[ob].apply(lambda x : len(x.split(",")))

new_df = df.pivot(index="date", columns="country", values="daily_vaccinations")
# new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
new_df = new_df[["Iceland", "Luxembourg"]]

fig, axs = plt.subplots(2,2, figsize=(15,10))

axs[0][1] = sns.lineplot(data=new_df, ax=axs[0][1])
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)

locator = plt.MaxNLocator(prune="both", nbins=6)

axs[0][0] = sns.lineplot(data=type_accines[["Iceland", "Luxembourg"]], ax=axs[0][0])
axs[0][0].xaxis.set_major_locator(locator)
axs[0][0].set_ylabel("number of vaccines")

axs[0][1].xaxis.set_major_locator(locator)
axs[0][1].set_ylabel("daily vaccinations")


new_df = df.pivot(index="date", columns="country", values="daily_vaccinations")
# new_df = new_df.interpolate(method="time")
new_df.fillna(0, inplace = True)
new_df = new_df.astype(int, inplace = True)
new_df = new_df[["United States", "China"]]

axs[1][0] = sns.lineplot(data=type_accines[["United States", "China"]], ax=axs[1][0])
axs[1][0].xaxis.set_major_locator(locator)
axs[1][0].set_ylabel("number of vaccines")

axs[1][1] = sns.lineplot(data=new_df, ax=axs[1][1])
axs[1][1].xaxis.set_major_locator(locator)
axs[1][1].set_ylabel("daily vaccinations")
```
